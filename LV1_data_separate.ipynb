## 데이터 불러오기

#### 본인 파일 경로를 기본 경로로 저장

```python
# 원본 데이터가 있는 기본 경로 저장.
filePath = "/content/drive/My Drive/forElse"
```

```python
import pandas as pd

# hanssem_product_images.UTF-8.csv
origin_data = pd.read_csv(filePath+"/hanssem_product_images_UTF8.csv")
origin_data.head()
```

## 데이터 구성 파악하기

```python
LV0_data= list(set(origin_data['LV0']))
LV1_data= list(set(origin_data['LV1']))
LV2_data= list(set(origin_data['LV2']))
print("LV0_data\t 총 ", len(LV0_data), "개", LV0_data)
print("=========="*3)
print("LV1_data\t 총 ", len(LV1_data), "개", LV1_data)
print("=========="*3)
print("LV2_data\t 총 ", len(LV2_data), "개", LV2_data)
print("=========="*3)
```

## LV1 데이터를 LV0_데이터이름 디렉토리에 저장

#### LV0_data 변경

```python
# LV0_data 속 데이터명 변경 '/' -> '+'
import re
LV0_plus=[]
for i in LV0_data:
  # LV0_data 변경
  level0 = i
  #print(level0)
  this_LV0_df = origin_data[origin_data['LV0'] == i]
  
  slash_loc =[]
  for a in re.finditer('/', level0):
    slash_loc.append(a.start())
  
  if len(slash_loc)==1:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:]
  if len(slash_loc)==2:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:slash_loc[1]]+'+'+i[slash_loc[1]+1:]
  
  LV0_plus.append(level0)
  
  
#print(LV0_data)
#print(LV0_plus)
LV0_data = LV0_plus[:]
print(LV0_data)
```

## LV1 csv 파일 생성

```python
import os

csvPath = filePath+"/csv"
```

```python
# LV0별 디렉토리에 세부 LV1 csv 파일 생성.

import re

for i in LV0_data:
  #this_LV0_df = origin_data[origin_data['LV0'] == i]
  LV0Path = csvPath+'/LV0_'+i
  this_LV0_df = pd.read_csv(LV0Path+"/LV0_"+i+".csv")
  
  print("LV0_data: ",i)
  for j in LV1_data:
    # j일때 데이터프레임
    this_LV1_df = this_LV0_df[this_LV0_df['LV1'] == j]
    if len(this_LV1_df) !=0:
      print("LV1_data: ",j,"\t",len(this_LV1_df), "개")

      # LV0Path에 LV1 csv 파일 생성 
      csvFileName = LV0Path+'/LV1_'+j+'.csv'
      print(csvFileName)
      this_LV1_df.to_csv(csvFileName, encoding ='utf-8-sig', index = False)
      print(this_LV1_df)
  print("==========="*3,"\n")
  
```
