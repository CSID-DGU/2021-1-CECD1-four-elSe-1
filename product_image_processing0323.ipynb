```python
# LV0별로 csv 파일 생성.
# 데이터명 변경 '/' -> '+'
import re
LV0_plus=[]
for i in LV0_data:
  # LV0_data 변경
  level0 = i
  #print(level0)
  this_LV0_df = origin_data[origin_data['LV0'] == i]
  #print(this_LV0_df.head())
  slash_loc =[]
  for a in re.finditer('/', level0):
    slash_loc.append(a.start())
  #print(slash_loc)
  if len(slash_loc)==1:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:]
  if len(slash_loc)==2:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:slash_loc[1]]+'+'+i[slash_loc[1]+1:]
  #print(i, level0)
  LV0_plus.append(level0)
  
  #print(this_LV0_df['LV0'])
  # df.loc[df.grades<50,'result']='fail'

  # LV0 데이터 변경
  this_LV0_df.loc[:, 'LV0'] = level0
  #print(this_LV0_df['LV0'])
  csvFileName = '/content/drive/My Drive/forElse/'+'LV0_'+level0+'.csv'
  print(csvFileName)
  this_LV0_df.to_csv(csvFileName)

print(LV0_data)
print(LV0_plus)
LV0_data = LV0_plus[:]
#this_LV0_df['LV0']
```
