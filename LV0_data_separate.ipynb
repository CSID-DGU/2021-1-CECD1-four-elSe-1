## 데이터 불러오기

#### 본인 파일 경로를 기본 경로로 저장

```python
# 원본 데이터가 있는 기본 경로 저장.
filePath = "/content/drive/My Drive/forElse"
```

```python
import pandas as pd

# hanssem_product_images.UTF-8.csv
origin_data = pd.read_csv(filePath+"/hanssem_product_images_UTF8.csv")
origin_data.head()
```

## 데이터 구성 파악하기

```python
LV0_data= list(set(origin_data['LV0']))
LV1_data= list(set(origin_data['LV1']))
LV2_data= list(set(origin_data['LV2']))
print("LV0_data\t 총 ", len(LV0_data), "개", LV0_data)
print("=========="*3)
print("LV1_data\t 총 ", len(LV1_data), "개", LV1_data)
print("=========="*3)
print("LV2_data\t 총 ", len(LV2_data), "개", LV2_data)
print("=========="*3)
```

## 데이터 정보를 저장할 새로운 경로 생성

파일 이름: csv

```python
import os

csvPath = filePath+"/csv"
if not os.path.isdir(csvPath):
  os.mkdir(csvPath)
```

# LV0별로 csv 파일 생성

```python
# LV0별로 새로운 디렉토리에 csv 파일 생성.
# 데이터명 변경 '/' -> '+'
import re
LV0_plus=[]
for i in LV0_data:
  # LV0_data 변경
  level0 = i
  #print(level0)
  this_LV0_df = origin_data[origin_data['LV0'] == i]
  print(this_LV0_df.head())
  slash_loc =[]
  for a in re.finditer('/', level0):
    slash_loc.append(a.start())
  #print(slash_loc)
  if len(slash_loc)==1:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:]
  if len(slash_loc)==2:
    level0 = i[0:slash_loc[0]]+'+'+i[slash_loc[0]+1:slash_loc[1]]+'+'+i[slash_loc[1]+1:]
  #print(i, level0)
  LV0_plus.append(level0)
  
  
  # 데이터프레임의 LV0값 변경
  this_LV0_df['LV0'] = level0
  print(this_LV0_df['LV0'])

  LV0Path = csvPath+'/LV0_'+level0
  if not os.path.isdir(LV0Path):
    os.mkdir(LV0Path)

  csvFileName = LV0Path+'/LV0_'+level0+'.csv'
  print(csvFileName)
  this_LV0_df.to_csv(csvFileName)

print(LV0_data)
print(LV0_plus)
LV0_data = LV0_plus[:]
```
